```python
from google.colab import drive
!pip install pdfplumber
import pdfplumber


from transformers import pipeline
import nltk
nltk.download('punkt')
from nltk.tokenize import sent_tokenize

```

##Extracting Text From PDF
First step here is to extract the text from the pdf. Below I am using pdf plumber because PyPDF2 had more OCR issues.

```python
#using pdf plumber got rid of weird whitespace issue

def extract_text_from_pdf(pdf_path, skip_pages=[]):
    text = ""
    with pdfplumber.open(pdf_path) as pdf:
        for i, page in enumerate(pdf.pages):
            if i in skip_pages:
                continue
            text += page.extract_text() + "\n"
    return text

# path from google dirve
pdf_path = '/content/drive/MyDrive/normal_people.pdf'

# skip table of contents and first few blank pages
skip_pages = list(range(8))


normal_people_text = extract_text_from_pdf(pdf_path, skip_pages= skip_pages)
```

```python
#can check here if the text is clean as expected.
print(normal_people_text[:500])
```

##Preprocessing the Text


*  Tokenisation - splitting the sentencies using NLTK
*   Print first 100 sentences to verify !







```python
sentences = sent_tokenize(normal_people_text)
```

```python
# Print the first 100 sentences to verify tokenization
for i, sentence in enumerate(sentences[:100]):
    print(f"Sentence {i+1}: {sentence}")
    print()
```

## Basic Emotion Annotation

```python

# emotion detection model
emotion_pipeline = pipeline("text-classification", model="bhadresh-savani/distilbert-base-uncased-emotion")

# annotate
labeled_data = [(sentence, emotion_pipeline(sentence)[0]['label']) for sentence in sentences]

#to verify
for i, (sentence, label) in enumerate(labeled_data[:5]):
    print(f"Sentence {i+1}: {sentence}")
    print(f"Label: {label}")
    print()
```
